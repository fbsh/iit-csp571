In order to compare the empirical values with the results, parametric and non parametric functions had to be calculated. The parametric functions; mean and standard deviation were calculated; and the non-parametric functions; median and histrogram/boxplot graphs were used. The parametric functions: average and standard deviation, inherently take on the assumption that the data follows a normal distribution. The standard deviation values calculated above for the four features, shows that Petal Length and Petal Width to not follow a normal distribution. Therefore, the results do not agree with the empirical values. Thus, a normal distribution should not be assumed for the two petal features. Many of the observed petals have a significant amount of small data points causing the data to be negatively skewed. Therefore, a histogram and boxplot are better for indicting the true nature of the data.

```r
df <- read.csv(file=url, header=FALSE, sep="", as.is =4&9, col.names= c("mpg","cylinders","displacement","horsepower","weight","acceleration", "model year", "origin", "car name"))
str(df)
```

```r
df$horsepower = as.numeric(df$horsepower)
```

```r
Horse_base <- df$horsepower
Horse_NA_med <- df$horsepower
```

```r
Horse_NA_med[which(is.na(Horse_base))] <-median(Horse_base, na.rm =TRUE)
Horse_NA_med
```

The original mean (104.4694) that was calculated for Horsepower, ignored the 7 values in the column with a NA (previously a “?”). Therefore, the number of observations used to calculate the mean was 398 - 7 = 391. Thus, when the median (93.5) was used to fill in the 7 NA values; the mean was slightly lowered to 104.304. This is because the 7 newly inserted 93.5 values very slightly, positely skewed the data to the right. However, this very minor shift is almost negligable because there are 398 total observations.

```r
library(tidyr) 
data(Boston) 
head(Boston)
```

```r
fit1 =lm(medv~lstat ,data=Boston )
confint (fit1)
```

```r
plot(predict (fit1), residuals (fit1))
```

```r
plot(Boston$lstat ,Boston$medv) 
abline (fit1)
```

```r
ggplot(Boston,aes(medv,lstat))+geom_point()+geom_smooth()
```

```r
predict (fit1 ,data.frame(lstat=c(5 ,10 ,15) ), interval ="confidence")
```

```r
predict (fit1 ,data.frame(lstat=c(5 ,10 ,15) ), interval ="prediction")
```

```r
fit2=lm(medv~lstat +I(lstat ^2),data=Boston) 
ggplot(Boston,aes(medv,lstat,I(lstat ^2)))+geom_point()+geom_smooth(method="lm", se=TRUE)+geom_point() + scale_color_discrete(guide=FALSE) + scale_fill_discrete(guide=FALSE)
```

```r
par(mfrow=c(2,2)) 
plot(fit2)
```

No prediction based on confidence and prediction are not the same Model 1 has only one predictor where as Model 2 has two different predictors. The compare the models we should perform anova which states the null hypothesis as two models fit equally well where as the alternative hypothesis states that the model 2 is a better fit. When we look at Fstatistics and p values it is obvious that model2 fits way better. Since we found a non linearity before, it was a signal that the second model will fit better.
