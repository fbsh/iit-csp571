```r
data(USArrests)
# scale dataset
USArrests <- scale(USArrests)
```

We choose to scale the data to avoid the cluster results dominated by certain features.

```r
# obtain the within-cluster sum of square for each k
within.ssq <- c()
k.vec <- 2:10
for(k in k.vec){
    km <- kmeans(USArrests, k)
    this.ssq <- km$tot.withinss
    within.ssq <- c(within.ssq, this.ssq)
}
```

```r
plot(k.vec, within.ssq, type='b')
```

Using Elbow's method, we can see that the optimal cluster can be given by 9.

```r
km.opt <- kmeans(USArrests, 9)
fviz_cluster(km.opt, USArrests, ellipse.type = "norm")
```
