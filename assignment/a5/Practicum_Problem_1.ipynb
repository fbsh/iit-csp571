```r
wineurl <- 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'
wine <- read.csv(wineurl, header=FALSE) 
colnames(wine) <- c('Type', 'Alcohol', 'Malic', 'Ash', 'Alcalinity', 'Magnesium', 'Phenols', 
                    'Flavanoids', 'Nonflavanoids', 'Proanthocyanins', 'Color', 'Hue', 
                     'Dilution', 'Proline')
```

```r
# apply pca
wine.pca <- prcomp(~.,wine, scale=T)
```

Scaling is used so that the result wouldn't be dominated by features with larger scale.

```r
biplot(wine.pca)
```

PC1 points to the negative direction of Hue. It suggests that the feature is negatively correlated with Hue.

```r
# supported value for PC1 with Hue
wine.pca$rotation['Hue','PC1']
```

```r
library(ggplot2)
# obtain variance
variance = wine.pca$sdev^2 / sum(wine.pca$sdev^2)
 
# Scree plot
qplot(c(1:length(variance)), variance) +
  geom_line() +
  geom_point(size=4)+
  xlab("Principal Component") +
  ylab("Variance Explained") +
  ggtitle("Scree Plot") +
  ylim(0, 0.5)
```

```r
# percentage of total variance explained by PC1 and PC2
variance[1:2]
```
