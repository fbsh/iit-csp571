## a

$P$ is the dimension of the space, and the $j$-th coordinate of the centroid for cluster $C_{k}$ is given by

$$
\bar{x}_{k, j}=\frac{1}{C_{k}} \sum_{i \in C_{k}} x_{i, j}
$$

For $1 \leq j \leq P$, we have:

$$
\begin{aligned}
\frac{1}{2\left|C_{k}\right|} \sum_{i, i^{\prime} \in C_{k}}\left(x_{i, j}-x_{i^{\prime}, j}\right)^{2} &=\frac{1}{2\left|C_{k}\right|} \sum_{i \in C_{k}} \sum_{i^{\prime} \in C_{k}}\left(x_{i, j}^{2}+x_{i^{\prime}, j}^{2}-2 x_{i, j} x_{i^{\prime}, j}\right) \\
&=\frac{1}{2}\left(\sum_{i \in C_{k}} x_{i, j}^{2}\right)+\frac{1}{2}\left(\sum_{i^{\prime} \in C_{k}} x_{i^{\prime}, j}^{2}\right)-\frac{1}{\left|C_{k}\right|}\left(\sum_{i, \in C_{k}} \sum x_{i^{\prime} \in C_{k}} x_{i, j} x_{i^{\prime}, j}\right) \\
&=\left(\sum_{i \in C_{k}} x_{i, j}^{2}\right)-\frac{1}{\left|C_{k}\right|}\left(\sum_{i \in C_{k}} x_{i, j}\right)\left(\sum_{i^{\prime} \in C_{k}} x_{i^{\prime}, j}\right) \\
&=\left(\sum_{i \in C_{k}} x_{i, j}^{2}\right)-\left|C_{k}\right| \bar{x}_{k, j}^{2} \quad(\text { using above equation}) \\
&=\left(\sum_{i \in C_{k}} x_{i, j}^{2}\right)-2\left|C_{k}\right| \bar{x}_{k, j}^{2}+\left|C_{k}\right| \bar{x}_{k, j}^{2} \\
&=\left(\sum_{i \in C_{k}} x_{i, j}^{2}\right)-2\left(\sum_{i \in C_{k}} x_{i, j}\right) \bar{x}_{k, j}+\left|C_{k}\right| \bar{x}_{k, j}^{2} \\
&=\sum_{i \in C_{k}}\left(x_{i, j}^{2}-2 x_{i, j} \bar{x}_{k, j}+\bar{x}_{k, j}^{2}\right) \\
&=\sum_{i \in C_{k}}\left(x_{i, j}-\bar{x}_{k, j}\right)^{2}
\end{aligned}
$$

$\frac{1}{\left|C_{k}\right|} \sum_{i, i^{\prime} \in C_{k}} \sum_{j=1}^{p}\left(x_{i j}-x_{i^{\prime} j}\right)^{2}=2 \sum_{i \in C_{k}} \sum_{j=1}^{p}\left(x_{i j}-\bar{x}_{k j}\right)^{2}$ (12.18) is proved.

The table gives the dissimilarity between classes, meaning that in order to produce clusters we need only select pairs whose values are small. The smallest value in the table is for the entries $(1,2)$ and $(2,1)$, followed by the pair $(3,4),(4,3)$

# 4a

Single and complete linkage use the minimal, and maximal distance respectively. Thus there are two possibilities either,
$$
D_{\min }^{\{1,2,3\},\{4,5\}}<D_{\max }^{\{1,2,3\},\{4,5\}}
$$
or,
$$
D_{\min }^{\{1,2,3\},\{4,5\}}=D_{\max }^{\{1,2,3\},\{4,5\}}
$$
where,
$$
D_{\min }^{\{1,2,3\},\{4,5\}}=\min \left(d_{14}, d_{15}, d_{24}, d_{25}, d_{34}, d_{35}\right)
$$
and,
$$
D_{\max }^{\{1,2,3\},\{4,5\}}=\max \left(d_{14}, d_{15}, d_{24}, d_{25}, d_{34}, d_{35}\right)
$$

but if $D_{\min }^{\{1,2,3\},\{4,5\}}=D_{\max }^{\{1,2,3\},\{4,5\}}$ then the points are equidistant. This implies that there would be a single cluster of $\{1,2,3,4,5\}$ instead of the two separate clusters.
Thus
$$
D_{\min }^{\{1,2,3\},\{4,5\}}<D_{\max }^{\{1,2,3\},\{4,5\}}
$$
is true. Consequently the clusters will fuse at a lower height in the single linkage dendogram.
