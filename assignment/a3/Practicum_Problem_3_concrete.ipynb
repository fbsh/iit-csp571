## P3

```r
concrete <- read_excel('Concrete_Data.xls')
concrete <- data.frame(concrete)
colnames(concrete) <- c("cem", "bfs", "fa", "water", "sp", "cagg", "fagg", "age", "ccs")
```

```r
# fit generalized additive model
# gam with linear terms
gam.linear <- gam(ccs~cem+fa+water+sp+cagg+fagg+age, data=concrete)
# r2
summary(gam.linear)$r.sq
```

```r
# gam with smoothed terms
gam.smooth <- gam(ccs~s(cem)+s(fa)+s(water)+s(sp)+s(cagg)+s(fagg)+s(age), data=concrete)
# r2
summary(gam.smooth)$r.sq
```

The r2 for gam with smoothed terms is higher than that wiht linear terms.

```r
# Visualize the regression with smoothed terms
visreg(gam.smooth)
```

The response variable typically centered around the prediction given by glm. However, for sp and fa with 0 value, the confidence interval didn't capture the most of the actual value given the large variance of the actual.
